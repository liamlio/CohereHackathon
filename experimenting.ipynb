```python
patent = {
    "Images": None,
    "Info": None,
    "Links": None,
    "Classifications": None,
    "Definitions": None,
    "Abstract": None,
    "Description": None,
    "Claims": None,
    "Applications": None,
    "Metadata": None,
    "ID": None,    
    "Published as": None,
    "Similar Documents": None,
    "Legal Events": None
}
for section in sections:
    if section.h2:
        for key in patent.keys():
            if key in str(section.h2):
                patent[key] = section
        print(section.h2)
    else:
        print(section)
        if "metadata" in str(section):
            patent["Metadata"] = section
            print(section)
```

```python
claims = "".join(list(patent["Claims"].find("div", {"itemprop": "content"}).strings)).replace("\n", "").split("  ")
```

```python
patent["Description"].find("div", {"itemprop": "content"}).strings
```

```python
description = "".join(list(patent["Description"].find("div", {"itemprop": "content"}))).replace("\n", "").split("  ")
description = [d for d in description if d != ""]
for i,des in enumerate(description):
    if len(des) < 50:
        description[i-1] += des
        description.pop(i)
description = [d.encode("utf-8").decode("utf-8") for d in description if d != ""]

description
```
